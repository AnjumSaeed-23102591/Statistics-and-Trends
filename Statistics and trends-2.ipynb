# Step 2: Plotting Functions

## 1. Plot 1: Histogram of Streams

```python
def plot_streams_histogram(data):
    # Grouping by released_year and summing streams for each year
    grouped_data = data.groupby('released_year')['streams'].sum().reset_index()

    plt.figure(figsize=(10, 6))
    plt.bar(grouped_data['released_year'], grouped_data['streams'], color='skyblue', edgecolor='black')
    plt.xlabel('Released Year')
    plt.ylabel('Total Streams')
    plt.title('Total Streams by Released Year')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.yscale('log')  # Using log scale if there are large discrepancies
    plt.show()

plot_streams_histogram(data)
```

## 2. Plot 2: Scatter Plot of Danceability vs. Energy

```python
def plot_danceability_energy(data):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x='danceability_%', y='energy_%', hue='released_year', palette='viridis', alpha=0.7)
    plt.xlabel('Danceability (%)')
    plt.ylabel('Energy (%)')
    plt.title('Danceability vs. Energy by Release Year')
    plt.show()

plot_danceability_energy(data)
```

## 3. Plot 3: Bar plot for Top 10 Most Streamed Songs

```python
def plot_top_streamed_songs(data):
    # Convert 'streams' to numeric, coercing errors to NaN
    data['streams'] = pd.to_numeric(data['streams'], errors='coerce')

    # Drop rows with NaN values in 'streams' if any exist
    data.dropna(subset=['streams'], inplace=True)

    # Sort the dataset by streams and select the top 10 songs
    top_streamed_songs = data.sort_values(by='streams', ascending=False).head(10)

    # Convert streams to millions
    top_streamed_songs['streams_millions'] = top_streamed_songs['streams'] / 1_000_000

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.barh(top_streamed_songs['track_name'], top_streamed_songs['streams_millions'], color='skyblue')
    plt.xlabel('Number of Streams (in Millions)')
    plt.title('Top 10 Most Streamed Songs')
    plt.gca().invert_yaxis()  # Invert y-axis to have the most streamed song on top
    plt.xticks(rotation=0)  # Ensure x-ticks are horizontal
    plt.show()

plot_top_streamed_songs(data)
```
